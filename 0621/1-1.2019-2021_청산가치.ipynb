```python
base_url = 'https://opendart.fss.or.kr/api/fnlttMultiAcnt.json'
crtfc_key = 'e734d2ae27450ddcc20a5a8eb5d46c107c28f88a'
bsns_year = input("년도수를 입력해 주세요(예시 :2021) : ")


a = input('몇 분기를 원하시나요? 숫자만 입력해주세요(예시 :1/2/3/4) : ') # a와 b는 임의값
if a == '1' :
    b = str(11013)
elif a == '2' :
    b = str(11012)
elif a == '3' :
    b = str(11014)
else :
    b = str(11011)
    
reprt_code = b


df_corpcode = pd.read_csv('Dart_codelist_last30_2.csv', usecols = ['종목명','종목코드','기업코드','수정일'], converters={'종목코드':str,'기업코드':str})


JM =  pd.DataFrame([], columns = ['종목명','분기','기업코드', '유동자산', '부채총계'])


if reprt_code[-1] == '3':
    Bungi = "1분기 보고서"
elif reprt_code[-1] == '2' :
    Bungi ="반기 보고서"
elif reprt_code[-1] == '4' :
    Bungi ="3분기 보고서"
else :
    Bungi = "사업보고서"

for y in df_corpcode.index: #인덱스 값으로 기업코드 찾고 그 찾은 기업코드를 request_url로 넣었습니다.
    try:
        corp_code_x = df_corpcode['기업코드'][y]
        name = df_corpcode['종목명'][y]

        request_url = base_url + '?' + 'crtfc_key='+crtfc_key+'&corp_code='+corp_code_x+\
        '&bsns_year='+bsns_year + '&reprt_code='+reprt_code

        response = requests.get(request_url).content # request_url을 넣어서 열리는 json파일을 요청하는 부분입니다!
        data = json.loads(response)
        data = data['list']
        df = pd.DataFrame(data)

        Money = int(df['thstrm_amount'][13].replace(',','')) #유동자산을 변수에 넣는 부분
        debt = int(df['thstrm_amount'][18].replace(',','')) #부채총계를 변수에 넣는 부분
        
        date = bsns_year + '년 ' + Bungi

        JM =  JM.append({'종목명' :name,'분기' : date, '기업코드' : corp_code_x,'유동자산' : Money,'부채총계' : debt},ignore_index=True) #JM(재무재표) 데이터 프레임에 각각의 값을 추가하는 부분입니다!!
    except:
        error_corp = name + ' : '+ corp_code_x
        print(error_corp)

JM['청산가치'] = JM['유동자산']-JM['부채총계']
```

```python
JM.to_csv("JM_2021_1Q.csv", header=True, index=False, encoding ='utf-8-sig')  
```
